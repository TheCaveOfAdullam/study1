```python
import numpy as np
import pandas as pd
from scipy.signal import spectrogram
import matplotlib.pyplot as plt
```

```python
# CSV 파일 경로 설정
csv_file_path = "newData2/STFMK-20201105-LW15-2055_20201125_163921_004.csv"
```

```python
# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)
```

```python
# 데이터 확인
print("Data Head:")
print(data.head())
```

```python
# 주파수 데이터 추출
frequencies = data['y'].values
```

```python
# 샘플링 주파수 설정
sampling_freq = 1000  # 샘플링 주파수 (예: 1000 Hz)}
```

```python
# STFT 계산
f, t, Sxx = spectrogram(frequencies, fs=sampling_freq, nperseg=256, noverlap=128)
```

```python
# 스펙트로그램 시각화
plt.figure(figsize=(10, 6))
plt.pcolormesh(t, f, 10 * np.log10(Sxx), shading='gouraud')
plt.title('Spectrogram of Frequency Data')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.colorbar(label='Intensity [dB]')
plt.ylim(0, 500)  # 주파수 범위 설정 (예: 0 ~ 500 Hz)
plt.tight_layout()
plt.show()
```
